## fasttext 词向量训练文件准备

```python
import pandas as pd
import numpy as np
train = pd.read_csv("../inputs/train.csv",quotechar='"')
vali = pd.read_csv("../inputs/vali.csv",quotechar='"')
test = pd.read_csv("../inputs/testa.csv",quotechar='"')
```

```python
allTextarr = np.concatenate([train.content.values, test.content.values, vali.content.values],0)
```

```python
allTextarr
```

```python
with open("../inputs/Source.txt", 'w') as out:
    def add_seg_line(line):
        print(line.strip('\"'),file=out)
    fv = np.vectorize(add_seg_line)
    _ = fv(allTextarr)
```
