# 长度特征

```python
train['num_words'] = train.content.str.count('\S+')
```

```python
train['num_sents'] = train.content.str.count('[。！？]')
```

```python
import numpy as np
import re
train['str_len'] = train.content.apply(lambda x: np.sum([len(w) for w in str(x).split()]))
train['mean_word_len'] = train.content.apply(lambda x: np.mean([len(w) for w in str(x).split()]))
train['chinese_word_count'] = train['content'].apply(lambda comment: sum(1 for c in comment.split() if
                                                       re.fullmatch(r'[\u4e00-\u9fff]+', c)))
train['num_unique_words'] = train['content'].apply(lambda comment: len(set(w for w in comment.split())))
train['words_vs_unique'] = train['num_unique_words'] / train['num_words']
train['chi_word_rate'] = train.apply(lambda row: float(row['chinese_word_count']) / float(row['num_words']),
                               axis=1)
```
